### Heroes Of Pymoli Data Analysis
-----
* There are a total of 586 players, 84% of whom are male. 14% are female.
* The average total purchase per person does not vary significantly by gender between players.
* The widest demographic for the game is 20-24 year olds (~45%), but there is a large variety of players, ranging from under 10 years old (3%) to over 40 (2%). Subsequently, this majority demographic also spent the most money on the game (4.40 USD average total purchase per person). 
* Interestingly, the <10 group was the next top spender, but this is likely a biased estimate due to the small sample size.
* The top spender in the population spent close to 19 USD on microtransactions. 
* The most popular weapon in the game (Oathbreaker) was purchased 12 times, netting a profit of ~50 USD. This made Oathbreaker the most profitable item in the game, in addition to the most popular. 

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load 
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file, encoding="UTF-8")
purchase_data.head() # test
```

## Player Count

* Display the total number of players


```python
totalplayers = len(purchase_data["SN"].unique())
totalplayers
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item ID"].unique())
unique_items
```

```python
average_price = round(purchase_data["Price"].mean(),2)
average_price
```

```python
purchase_counts = len(purchase_data["Item Name"])
purchase_counts
```

```python
total_revenue = purchase_data["Price"].sum()
total_revenue
```

```python
analysis_df = pd.DataFrame({"Number of Unique Items": unique_items, "Average Price": average_price,
                          "Number of Purchases": purchase_counts, "Total Revenue": [total_revenue]})
analysis_df.head()
```

```python
# format table
analysis_df["Average Price"] = analysis_df["Average Price"].map("${:.2f}".format)
analysis_df["Total Revenue"] = analysis_df["Total Revenue"].map("${:.2f}".format)
analysis_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_count_df = purchase_data.groupby("Gender")["SN"].nunique()
gender_count_df.head()
gender_percent = round((gender_count_df / totalplayers * 100),2)
gender_percent.head()
gender_demographics = pd.DataFrame({"Total Count": gender_count_df,"Percentage of Players":gender_percent})
gender_demographics.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_group = purchase_data.groupby("Gender")["Purchase ID"]
purchase_count = purchase_group.count()
gender_purchase_total = purchase_data.groupby(["Gender"])["Price"].sum()
gender_average = round(purchase_data.groupby(["Gender"])["Price"].mean(),2)
gender_counts = purchase_data.groupby("Gender")["Price"].count()
avg_total_per_person = round(gender_purchase_total / gender_demographics["Total Count"],2)

gender_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                                "Average Purchase Price": gender_average,
                                "Total Purchase Value": gender_purchase_total,
                                "Avg Total Purchase per Person": avg_total_per_person})
gender_analysis.head()
```

```python
# format table
gender_analysis["Average Purchase Price"] = gender_analysis["Average Purchase Price"].map("${:.2f}".format)
gender_analysis["Avg Total Purchase per Person"] = gender_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)
gender_analysis.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
drop_duplicates = purchase_data.drop_duplicates(["SN"])
age_group_count = drop_duplicates.groupby("Age Group").count()
age_percent = (age_group_count["Purchase ID"]/totalplayers*100).map("{:,.2f}".format)
age_demographics = pd.DataFrame({"Total Count": age_group_count["Purchase ID"], "Percentage of Players": age_percent})
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_group_age = purchase_data.groupby("Age Group")["Purchase ID"].count()
age_purchase_total = purchase_data.groupby("Age Group")["Price"].sum()
age_average_purchase = purchase_data.groupby("Age Group")["Price"].mean()
age_average_perperson = age_purchase_total / (age_group_count["Purchase ID"])
age_analysis = pd.DataFrame({"Purchase count": purchase_group_age, 
                            "Average Purchase Price": age_average_purchase,
                            "Total Purchase Value": age_purchase_total,
                            "Avg Total Purchase per Person": age_average_perperson})
age_analysis.head()
```

```python
# formatting
age_analysis["Average Purchase Price"] = age_analysis["Average Purchase Price"].map("${:.2f}".format)
age_analysis["Total Purchase Value"] = age_analysis["Total Purchase Value"].map("${:.2f}".format)
age_analysis["Avg Total Purchase per Person"] = age_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)
age_analysis.head()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_count_spenders = purchase_data.groupby("SN")["Purchase ID"].count()
average_purchase_spenders = purchase_data.groupby("SN")["Price"].mean()
total_purchase_spenders = purchase_data.groupby("SN")["Price"].sum()
spender_analysis = pd.DataFrame({"Purchase Count": purchase_count_spenders,
                                "Average Purchase Price": average_purchase_spenders,
                                "Total Purchase Value": total_purchase_spenders})
spender_analysis = spender_analysis.sort_values("Total Purchase Value", ascending=False)
spender_analysis.head()
```

```python
# formatting
spender_analysis["Total Purchase Value"] = spender_analysis["Total Purchase Value"].map("${:.2f}".format)
spender_analysis.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_group_count = purchase_data.groupby(["Item ID", "Item Name"]).count()
popular_group = purchase_data.groupby(["Item ID", "Item Name"])
popular_count = popular_group_count["Purchase ID"]
item_price = popular_group["Price"].mean()
popular_total = popular_group["Price"].sum()

popular_analysis = pd.DataFrame({"Purchase Count": popular_count,
                                "Item Price": item_price,
                                "Total Purchase Value": popular_total})
popular_sorted = popular_analysis.sort_values("Purchase Count", ascending=False)
popular_sorted.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable = popular_sorted.sort_values("Total Purchase Value", ascending=False)
most_profitable.head()
```

```python
# formatting
popular_sorted["Item Price"] = popular_sorted["Item Price"].map("${:.2f}".format)
popular_sorted["Total Purchase Value"] = popular_sorted["Total Purchase Value"].map("${:.2f}".format)
popular_sorted.head()
```

```python
# formatting
most_profitable["Item Price"] = most_profitable["Item Price"].map("${:.2f}".format)
most_profitable["Total Purchase Value"] = most_profitable["Total Purchase Value"].map("${:.2f}".format)
most_profitable.head()
```
